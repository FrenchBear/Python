```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
from scipy import linalg
import math
%matplotlib inline
%config InlineBackend.figure_format='svg'
```

```python
# La fréquence (densité de probabilité) des valeurs propres d'une matrice aléatoire symmétrique (distribution normale) se répartit en demi-cercle
# Version normalisée (en divisant par racine(N)) de façon à tomber dans l'intervale -2..2
N = 1000
A = np.random.randn(N, N)/np.sqrt(N)
M = np.tril(A) + np.tril(A, -1).T
eigvals = np.real(linalg.eigvals(M))
n, bins, patches = plt.hist(eigvals, bins=50)
```

```python
# Histogramme des écarts successifs entre les valeurs propres d'une matrice aléatoire symmétrique (loi normale)
# Limite = courbe GOE, Gaussian Orthogonal Ensemble
diffeigvals = np.diff(sorted(eigvals))
n, bins, patches = plt.hist(diffeigvals, bins=50)
```
