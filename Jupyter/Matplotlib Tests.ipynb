```python
#!/usr/bin/env python

# implement the example graphs/integral from pyx
#from pylab import *
from matplotlib.patches import Polygon

def func(x):
    return (x-3)*(x-5)*(x-7)+85

ax = plt.subplot(111)

a, b = 2, 9 # integral area
x = np.arange(0, 10, 0.01)
y = func(x)
plt.plot(x, y, linewidth=1)

# make the shaded region
ix = np.arange(a, b, 0.01)
iy = func(ix)
verts = [(a,0)] + list(zip(ix,iy)) + [(b,0)]
poly = Polygon(verts, facecolor='0.8', edgecolor='k')
ax.add_patch(poly)

plt.text(0.5 * (a + b), 30,
     r"$\int_a^b f(x)\mathrm{d}x$", horizontalalignment='center',
     fontsize=20)

plt.axis([0,10, 0, 180])
plt.figtext(0.9, 0.05, 'x')
plt.figtext(0.1, 0.9, 'y')
ax.set_xticks((a,b))
ax.set_xticklabels(('a','b'))
ax.set_yticks([])
```

```python
# larger image, thick lines
plt.figure(figsize=(10,5), dpi=80)
plt.subplot(111)
X = np.linspace(-np.pi, np.pi, 256,endpoint=True)
C,S = np.cos(X), np.sin(X)
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label="Cosine")
plt.plot(X, S, color="red", linewidth=2.5, linestyle="-", label="Sine")

xmin, xmax = X.min(), X.max()
ymin, ymax = C.min(), C.max()
dx = (xmax - xmin) * 0.05
dy = (ymax - ymin) * 0.05
plt.xlim(xmin - dx, xmax + dx)
plt.ylim(ymin - dy, ymax + dy)

ax = plt.gca()     # gca = Get Current Axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')

ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))

#xticks(np.linspace(-3,3,13,endpoint=True))
#yticks(np.linspace(-1,1,5,endpoint=True))

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],[r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, +1], [r'$-1$', '$+1$'],rotation=30)

plt.legend(loc='upper left')   # Or use 'best'

t = 2*np.pi/3
plt.plot([t,t],[0,np.sin(t)], color ='red', linewidth=1.5, linestyle="--")
plt.scatter([t,],[np.sin(t),], 50, color ='red')     # Big red dot
plt.annotate(r'$\sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$', xy=(t, np.sin(t)), xycoords='data',
    xytext=(+10, +30), textcoords='offset points', fontsize=16,
    arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plt.plot([t,t],[0,np.cos(t)], color ='blue', linewidth=1.5, linestyle="--")
plt.scatter([t,],[np.cos(t),], 50, color ='blue')    # Add a big dot
plt.annotate(r'$\cos(\frac{2\pi}{3})=-\frac{1}{2}$', xy=(t, np.cos(t)), xycoords='data',
    xytext=(-90, -50), textcoords='offset points', fontsize=16,
    arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(16)
    label.rotation = 10
    label.set_bbox(dict(facecolor='yellow', edgecolor='None', alpha=0.65 ))
```

```python
# fill_between example
n = 256
X = np.linspace(-np.pi,np.pi,n,endpoint=True)
Y = np.sin(2*X)
plt.plot(X, Y+1, color='blue',alpha=1.00)
plt.plot(X, Y-1, color='blue',alpha=1.00)
plt.fill_between(X, Y+1, 1, color='blue', alpha=0.25)
plt.fill_between(X, Y-1, -1, color='blue', alpha=0.25, where=Y>0)
plt.fill_between(X, Y-1, -1, color='red', alpha=0.25, where=Y<0)
_ = plt.xticks([]), plt.yticks([])
```

```python
# bar plot example
n = 12
X = np.arange(n)
Y1 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)
Y2 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)
plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')
for x,y in zip(X,Y1):
    plt.text(x, y+0.02, '%.2f' % y, ha='center', va= 'bottom', size=8)
for x,y in zip(X,-Y2):
    plt.text(x, y-0.04, '%.2f' % (-y), ha='center', va= 'top', size=8)
plt.xticks([]), plt.yticks([])
_ = plt.ylim(-1.25,+1.25)
```

```python
# controur plot example
def f(x,y): return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)
n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)
plt.contourf(X, Y, f(X,Y), 8, alpha=.75, cmap='hot')
C = plt.contour(X, Y, f(X,Y), 8, colors='black', linewidths=.5)
plt.xticks([]), plt.yticks([])
_ = plt.clabel(C,fontsize=8,inline=True)
```

```python
# vetor plot example (quiver)
n = 9
X,Y = np.mgrid[0:n,0:n]                      # X,Y represent vector origin
R = 10+np.sqrt((Y-n/2.0)**2+(X-n/2.0)**2)    # R = distance from center, used to select color
T = np.arctan2(Y-n/2.0, X-n/2.0)             # Angle at point of coordinate [X,Y] = vector[X,Y]
U,V = R*np.cos(T), R*np.sin(T)               # U,V = actual vectors (x coord, y coord) to draw
#axes([0.025,0.025,0.95,0.95])
plt.quiver(X,Y,U,V,R, alpha=.5)
plt.quiver(X,Y,U,V, edgecolor='k', facecolor='None', linewidth=.5)
plt.xlim(-1,n), plt.xticks([])
_ = plt.ylim(-1,n), plt.yticks([])
```
